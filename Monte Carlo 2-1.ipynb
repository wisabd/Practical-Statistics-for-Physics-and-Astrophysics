# Inverse Transform

```python
#Import the necessary libraries and functions
from math import sqrt
from math import *
import random
import matplotlib.pyplot as plt
import numpy as np

X=[] #Setting the lists to store the values of x and y 
Y=[]

#Inverse Tranform
#To find the inverse factor for this case: 
#
#

for i in range(10**6): # Loop for producing 10^6 points 
    r=3*sqrt(np.random.random()) #Generating the random variables according to the transform
    phi=2*pi*np.random.random()
    x=r*np.cos(phi) #Changing to cartesian coordinates
    y=r*np.sin(phi)
    X.append(x) #Appending the cartesian coordinates to the corresponding list
    Y.append(y)

#Plot Distribution of Points
Xmin=np.min(X) #To get the max and min values for the range of the plot
Ymin=np.min(Y)
Xmax=np.max(X)
Ymax=np.max(Y)
x_bins=np.linspace(Xmin, Xmax, 60) #Generating the 0.1x0.1 bins  as 6/60=0.1
y_bins=np.linspace(Ymin, Xmax, 60)
fig, ax = plt.subplots(figsize=(6,6))
ax.hist2d(X, Y, bins=[x_bins, y_bins]) #cmap=plt.cm.nipy_spectral) #Plotting the 2D Histrogram
ax.set_title('2D Distribution by Inverse Transform') #Settigns of the figure
ax.set(xlabel='x axis',ylabel='y axis')
#bar = fig.colorbar() #Tried to add a color bar; didn't work
#ax.colorbar()
fig.savefig('invtransf.png')
```
