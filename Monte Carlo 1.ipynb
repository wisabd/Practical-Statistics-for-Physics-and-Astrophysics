# Pi Calculation

```python
#Import the necessary libraries and functions
from math import sqrt
from math import *
import random
import matplotlib.pyplot as plt
import numpy as np

inside=0 #Setting up the counters
outside=0
pi_values=[] #Setting up the list to store the values 
pi_errors=[]
pi_rel=[]
points=[100,500,1000,5000,10000,50000,100000,500000,1000000,5000000,10000000] #The number of points for each run
sqrtN=[] # Setting the list to store the 1/sqrt(N) values

#Pi calculation
for samples in points: #This loop to add each value of 1/sqrt(N) for plotting
    u= 1/sqrt(samples)
    sqrtN.append(u)
    for i in range(samples): #This loop for \pi calculation
        x=random.random() #Generating the random numbers
        y=random.random()
        
        if sqrt(x*x+y*x)<=1: #Checking whether the point falls outside/inside the section
            inside +=1
        else:
            outside +=1
            
    piv = 4.0 * ((inside)/(inside+outside)) #Values of \pi are four times the ratio inside/total
    error = pi-piv #Absolute error respect to real \pi
    rel =(pi-piv)/pi #Relative error respect to real\pi
    pi_values.append(piv) #Appending the results of each round to the respective lists
    pi_errors.append(error)
    pi_rel.append(rel)
    inside =0 #Resetting the value of the counters to zero after each round of points
    outside =0
    
#Plotting the results
#PART A) Plot Error vs. N
fig1, ax= plt.subplots(figsize=(12,5))
ax.plot(points,pi_errors)
ax.set(xlabel='Number of points',ylabel='Error of \u03C0', title='Error of \u03C0 vs. N')
fig1.savefig("Errors.png")
#PART B) Plot Relative error vs. 1/sqrt(N)
fig2, ax= plt.subplots(figsize=(12,5))
ax.plot(sqrtN,pi_rel)
ax.set(xlabel='1/sqrt(N)',ylabel='Relative error of \u03C0', title='Relative error of \u03C0 vs. 1/sqrtN')
fig2.savefig("Relative.png")
```
